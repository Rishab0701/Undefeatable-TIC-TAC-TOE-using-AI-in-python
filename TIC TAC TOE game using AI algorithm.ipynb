```python
import random # for easy level of opponent
import time # just for loading screen



def show(): # to display the board in a more organised manner without commas
    for i in board:
        print('â”‚',end='')
        for j in i:
            print(j,end='â”‚')
        print('\n')

def entervalidinput(giveninput,inputrange): # makes sure no wrong input is given by user
    if giveninput.isnumeric() and int(giveninput) in inputrange:
        return int(giveninput)
    
    else:
        giveninput = input("Enter a vaild input ")
        return entervalidinput(giveninput,inputrange)


    
def heuristic_value(board) :
    for i in range(3) :
        if (board[i][0]==board[i][1]==board[i][2]) or (board[0][i]==board[1][i]==board[2][i]):
            if (board[i][i] == player) or (board[i][i] == player) :
                return -10
            elif (board[i][i] == player2) or (board[i][i] == player2) :
                return 10

    if (board[0][0] == board[1][1]== board[2][2]) or (board[0][2] == board[1][1] == board[2][0]):
        
        if (board[1][1] == player) :
            return -10
        elif (board[1][1] == player2) :
            return 10
    return 0
    
def play(ip,who): # plays the move with given input,who's move this is
    who = who
    available_moves.remove(ip) # updates available moves for the next player
    
    # To covert keyboard inputs into index values so that values can be inserted in the board
    if ip in [7,8,9]:
        a = ip - 7
        board[0][a] = who
        
    elif ip in [4,5,6]:
        b= ip - 4
        board[1][b] = who
        
    elif ip in [1,2,3]:
        c= ip -1
        board[2][c] = who
        
def isMovesLeft(board) : #to check if the board is full or there is a block left
    for i in range(3) :
        for j in range(3) :
            if (board[i][j] == '   ') :
                return True
    return False


def pc_turn():
    
    if mode == 0:
        print('\n\n',player2,'\n AI "S TURN\n')
        index = bestmove(board)
        x,y =index
        board[x][y] = player2
        
        if  x==0 : vaL = 7+y
        elif x==1: vaL = 4+y
        elif x==2: vaL = 1+y
        
        available_moves.remove(vaL)
                
    else:
        print('\n\n',player2,'\n Computer "S TURN\n')
        pc_ip = random.choice(available_moves)
        pc_val = play(pc_ip,player2)
    show()
   
    
def user_turn(player):
    print('\n\n',player,'"s TURN\n')
    user_ip = (input('Enter the location'))
    user_ip = entervalidinput(user_ip,available_moves)
    user_val = play(user_ip,player)
    show()
    
def minimax(board,depth,isMax):
    if isMovesLeft(board)==False:
        return 0
    score = heuristic_value(board)
    if score == 10:
        return score
    if score == -10:
        return score
    
    if isMax:
        bestvalue = -1000 #The maximum possible negative value
        
        #Traverse all cells to evaluate all possiblity
        for i in range(3):
            for j in range(3):
                if board[i][j]=='   ':
                    board[i][j]=player2
                    best = minimax(board,depth+1,not isMax)
                    
                    bestvalue = max(best,bestvalue)
                    board[i][j]='   ' #undoing the move
        return bestvalue
    
    else:
        bestvalue = 1000 #The maximum possible positive value
        
        # Similar to above
        for i in range(3):
            for j in range(3):
                if board[i][j]=='   ':
                    board[i][j]=player
                    best = minimax(board,depth+1,not isMax)
                    
                    bestvalue = min(best,bestvalue)
                    board[i][j]='   '
        return bestvalue

def bestmove(board):
    bestvalue = -1000 # as we need max we initalize 
    bestmove = (-1,-1) #initalizing a variable
    
    for i in range(3):
        for j in range(3):
            if board[i][j] == '   ': #traversing all
                board[i][j] = player2
                value = minimax(board,1,False)
                board[i][j] = '   '
                
                if value>bestvalue: # as player 2 ('O') is the max player we need max value for move
                    bestmove = (i,j)
                    bestvalue = value
    return bestmove

def loadingscreen():
    print('. . . . . â•°â”€â”€â•®TIC TAC TOEâ•­â”€â”€â•¯ . . . . .')
    print('                         Coded byâ”ˆ á’ÉªsÊœá´€Ê™')
    print('  Loading Game....')
    print('âŒ© ',end=' ')
    for i in range(20):
        option = [0.123,0.100,0.03,0.09,0.1]
        t =random.choice(option)
        time.sleep(t)
        print('ã€‹',end= ' ')
    print('âŒª')
    
def choosestyle():
    print ('\n\nChoose what symbols you want in the game (Note: You will always make the first move)\n  ')
    print('PRESS         SYMBOLS')
    print('   0          â ğŸ…¾ï¸\n   1          â„ï¸ ğŸŒ\n   2          âš”ï¸ ğŸ›¡ï¸\n   3          ğŸ˜µ ğŸ˜¶\n\n')
    symbol = entervalidinput(input(),[0,1,2,3])
    available_styles = [['â','ğŸ…¾ï¸'],['â„ï¸','ğŸŒ'],['âš”ï¸','ğŸ›¡ï¸'],['ğŸ˜µ','ğŸ˜¶']]
    player, player2 =available_styles[symbol]
    return player,player2
        

# START OF GAME

loadingscreen()                    
print("\n\nWelcome to the game of TIC TAC TOE \n\nPress any key to begin the game\nPress 1 for instructions\n\n")
status = input()
mode=0
if status == '1':
    print('\nThe following are the position values of the table enter these valuesâŒ¨ to insert X in the given position\n\n 7  8  9 \n 4  5  6 \n 1  2  3 ')
    print('Use Numpad for a better experience')
status ='1'
inp=input("\n\nHUMAN vs AI (HARD- undefeatable)   --Press 0\nHUMAN vs COMPUTER (Very EASY)      --Press 1 \nHUMAN vs HUMAN (Play with a friend)--Press 2\n\n")
mode = entervalidinput(inp,[0,1,2])
inp = input('\n\nTo choose style press 0 \nPress any other key to continue..')
if inp=='0': player,player2 = choosestyle()
else: player,player2 = 'X','O'



while status == '1':
    board = [['   ','   ','   '],['   ','   ','   '],['   ','   ','   ']]
    available_moves = [1,2,3,4,5,6,7,8,9]
    show()
    
    while True:
        
        user_turn(player)
        if heuristic_value(board) == -10:
            print('Player wins')
            break
        elif isMovesLeft(board) == False:
            print('Its a Draw')
            break
            
        if mode==2:
            user_turn(player2)
            if heuristic_value(board) == 10:
                print('Player 2 wins')
                break
            elif isMovesLeft(board) == False:
                print('Its a Draw')
                break
            
        else:
            pc_turn()
            if heuristic_value(board) == 10:
                print('PC wins')
                break
            elif isMovesLeft(board) == False:
                print('Its a Draw')
                break
    
    
    status = input('Wanna a REMATCH enter 1 OR \nAfraid of LOSING press any other key to quit\n')
```
